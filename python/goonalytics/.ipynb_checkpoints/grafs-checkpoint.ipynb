```python
"""
graphs and shit
"""
from typing import List, Dict, Set

from goonalytics.io.gcloudio import BigQueryer
from goonalytics.settings import GCLOUD_TABLE_NAME as tbl
import networkx as nx
import matplotlib.pyplot as plt
import logging
import random
logging.basicConfig()
logger = logging.getLogger(__name__)
logger.setLevel(logging.DEBUG)


def get_post_quote_dict(thread_id: int) -> Dict[int, Set[int]]:
    mp = dict()
    post_user_dict = dict()
    bq = BigQueryer()
    qr = bq.sql_inject('''select post_id, user_id, quoted_post_ids from %s where thread_id=%d''' % (tbl, thread_id))
    for row in qr.rows:
        post_id = row[0]
        user_id = row[1]
        quote = row[2] if row[2] is not None else 'null'
        if post_id not in mp.keys():
            mp[post_id] = set()
        mp[post_id].add(quote)
        post_user_dict[post_id] = user_id
    # next transform the post_ids to the user ids
    logger.debug("Test case: %s", mp[458975822])
    mp_intermediate = dict()
    for key in mp.keys():
        user = post_user_dict[key]
        quote_users = set([post_user_dict[val] for val in mp[key] if val is not 'null'])
        mp_intermediate[user] = quote_users
    return mp_intermediate

def create_graph(user_dict: Dict[int, Set[int]]) -> nx.Graph:
    g = nx.Graph();
    # add all vertices
    for u in user_dict.keys():
        g.add_node(u, weight=random.random())
    for u in user_dict.keys():
        for u2 in user_dict[u]:
            g.add_edge(u, u2)
    return g

mp = get_post_quote_dict(3763968)
G = create_graph(mp)
# nx.draw(G)
# copied basically all this shit from examples
plt.figure(figsize=(8,8))
import random
colors = [random.random() for n in range(nx.number_of_nodes(G))]
nx.draw(G,node_size=80,node_color=colors,vmin=0.0,vmax=1.0)
plt.xlim(-0.15,1.2)
plt.ylim(-0.25,1.2)
plt.show()
```
