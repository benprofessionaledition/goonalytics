```python
from goonalytics.grafkl import get_post_quote_dict, create_graph
import matplotlib.pyplot as plt
import networkx as nx

mp, unames = get_post_quote_dict(3763968)
G = create_graph(mp)
# nx.draw(G)
# copied basically all this shit from examples
plt.figure(figsize=(8,8))
import random
colors = [random.random() for n in range(nx.number_of_nodes(G))]
nx.draw(G,node_size=80,node_color=colors,vmin=0.0,vmax=1.0,labels=unames)
plt.xlim(-0.15,1.2)
plt.ylim(-0.25,1.2)
plt.show()
```
